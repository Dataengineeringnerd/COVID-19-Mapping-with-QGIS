```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

# Read the shapefile
world = gpd.read_file("ne_50m_admin_0_countries_lakes/ne_50m_admin_0_countries_lakes.shp")

# Read the shapefile
world2 = gpd.read_file("ne_50m_admin_0_countries_lakes/ne_50m_admin_0_countries_lakes.shp")
```

```python
#Read the CSV file with COVID-19 data
covid_cases = pd.read_csv("time_series_covid19_confirmed_global.csv")
covid_cases.head(1)

# Group by Country/Region and sum the cases
covid_cases_grouped = covid_cases.groupby('Country/Region').sum()
#covid_cases_grouped.head()

# Merge COVID-19 data with world data
world = world.merge(covid_cases_grouped, how='left', left_on='ADMIN', right_on='Country/Region')

# Plot the map
ax = world.plot(column='1/26/22', cmap='Reds', legend=True, figsize=(15, 10))


# Add title and labels
plt.title('COVID-19 Cases by Country')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the map
plt.show()
```

```python
#Read the CSV file with COVID-19 data
covid_death = pd.read_csv("time_series_covid19_deaths_global.csv")


# Group by Country/Region and sum the cases
covid_death_grouped = covid_death.groupby('Country/Region').sum()

# Merge COVID-19 data with world data
world2 = world2.merge(covid_death_grouped, how='left', left_on='ADMIN', right_on='Country/Region')
#world.columns()
# Plot the map
ax = world2.plot(column='1/26/22', cmap='Reds', legend=True, figsize=(15, 10))

# Add title and labels
plt.title('COVID-19 Deaths by Country')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the map
plt.show()
```
